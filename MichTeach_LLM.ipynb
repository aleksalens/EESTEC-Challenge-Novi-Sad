```python
!pip install fastapi nest-asyncio pyngrok uvicorn

from fastapi import FastAPI
from fastapi.middleware.cors import CORSMiddleware
from fastapi import HTTPException, WebSocketException

app = FastAPI()

app.add_middleware(
    CORSMiddleware,
    allow_origins=['*'],
    allow_credentials=True,
    allow_methods=['*'],
    allow_headers=['*'],
)

from pydantic import BaseModel


class PageRequest(BaseModel):
    response: str

@app.post("/questions")
async def transform_to_json(page_request: PageRequest):
    try:
        # Extract the pageS string from the request body
        pageS = page_request.response
        print(pageS)
        # Open a new text file in write mode
        with open('document.txt', 'w') as file:
          file.write(pageS)
        document=[]
        loader=TextLoader('document.txt')
        document.extend(loader.load())
        # Store the filename in a variable
        # document = 'document.txt'
        print("-----")

        document_splitter=CharacterTextSplitter(separator='\n', chunk_size=24, chunk_overlap=8)

        document_chunks=document_splitter.split_documents(document)

        embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')

        vectordb=Chroma.from_documents(document_chunks,embedding=embeddings, persist_directory='./data')

        vectordb.persist()

        pipe=pipeline("text-generation",
              model=model,
              tokenizer=tokenizer,
              torch_dtype=torch.bfloat16,
              device_map='auto',
              max_new_tokens=128,
              min_new_tokens=-1,
              top_k=3

        )

        llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0.1})

        memory=ConversationBufferMemory(memory_key='chat_history', return_messages=True)

        pdf_qa=ConversationalRetrievalChain.from_llm(llm=llm,
                                             retriever=vectordb.as_retriever(search_kwargs={'k':2}),
                                             verbose=False, memory=memory)



        # System prompt to guide the model's response style
        system_prompt = """You are working as an assistant. \
        You always answer concisely. Just give me the answers, skip your introductory part.

        """
        # Creating a function to use the pipeline with system prompt
        def get_answer(user_prompt):
            full_prompt = system_prompt + user_prompt  # Concatenating system prompt with user prompt
            result =pdf_qa({"question": full_prompt})
            return result['answer']

        def get_llm_answer(user_prompt):
            full_prompt = system_prompt + user_prompt  # Concatenating system prompt with user prompt
            result = pipe(full_prompt)
            answer = result[0]['generated_text'].strip()
            return answer

        generated_questions=get_answer("Generate 5 questions about arthritis.")
        print(generated_questions)

        import re

        pattern = r'.*?1\.'

        match = re.search(pattern, generated_questions)
        if match:
            result = generated_questions[match.start():]
            print(result)
        else:
            print("Pattern not found")

        # Your transformation logic goes here
        # For now, let's just wrap the string in a dictionary
        transformed_json = {"response": result}

        # Return the transformed JSON
        return transformed_json
    except Exception as e:
        print(e)
        transformed_json = {"response": result}
```
